```python
censo2023 = pd.read_csv('./Inputs/Censo_CCAA_Provincias.csv', header=0, delimiter = ';')
censo2023
```

```python
aux = []
for provincia in provincias:
    #print (provincia)
    if provincia == 'TOTAL ESPAÑA': 
        continue
    row = censo2023[(censo2023['Provincias'] == provincia) & (censo2023['Año'] == 2023) & (censo2023['Sexo'] == 'Ambos sexos')].index
    #print (row)
    aux1 = censo2023['Total'][row[0]].replace('.', '')
    #print (aux1)
    aux.append(int(aux1))
aux.append(int(censo2023['Total'][0].replace('.', '')))
poblaciones['2023'] = aux
#poblaciones
```

```python
codigos = []

for i, provincia in enumerate(provincias):
    if provincia == 'TOTAL ESPAÑA': 
        continue
    codigos.append(provincia[:2])
    provincias[i] = provincia[3:]

codigos.append('00')
codigos
```

```python
coordenadas = pd.read_csv('./Inputs/Coordenadas_provincias.csv', header = 0, delimiter = ';')
coordenadas
```

```python
def weighted_coord (lat,long,poblacion):
    auxpob = poblacion['2023'][:-1]
    auxlat = [c*w for c,w in zip(lat,auxpob)]
    auxlong = [c*w for c,w in zip(long,auxpob)]
    return sum(auxlat)/sum(auxpob),sum(auxlong)/sum(auxpob)
```

```python
lat, long = [], []
for cod in codigos:
    if cod == '00':
        lat00,long00 = weighted_coord(lat,long,poblaciones)
        print (np.mean(lat), lat00, np.mean(long), long00)
        lat.append(lat00)
        long.append(long00)
    else:
        row = coordenadas[coordenadas['Codigo'] == int (cod)].index
        #print (row)
        indice = coordenadas['Geo Point'][row[0]].find(',')
        #print (indice)
        auxlat = float (coordenadas['Geo Point'][row[0]][:indice])
        auxlong = float (coordenadas['Geo Point'][row[0]][indice+1:])
        #print (auxlat, auxlong)
        lat.append(auxlat)
        long.append(auxlong)
    
```

```python
poblacion_provincias = pd.DataFrame()
poblacion_provincias ['Codigos'] = codigos
poblacion_provincias ['Provincias'] = provincias

poblacion_provincias ['Latitud'] = lat
poblacion_provincias ['Longitud'] = long

for año in range (2014,2024):
    poblacion_provincias[str(año)] = poblaciones[str(año)]
    
poblacion_provincias
```

```python
poblacion_provincias.to_csv('Cities_unified.csv', index=False)
```
