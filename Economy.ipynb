```python
import pandas as pd
import numpy as np
import re
```

```python
ipc = pd.read_csv('./Inputs/IPC_Provincias.csv', header = 0, delimiter = ';')
ipc
```

```python
ipc = ipc[ipc['Grupos ECOICOP'] == 'Índice general']
ipc = ipc[ipc['Tipo de dato'] == 'Índice']
ipc.reset_index(drop=True, inplace=True)
ipc
```

```python
año = []
for i in range (len(ipc)):
    #print (ipc['Periodo'][i])
    aux = ipc['Periodo'][i] [5:] + '-' + ipc['Periodo'][i][:4]
    año.append(int(ipc['Periodo'][i][:4]))
    ipc['Periodo'][i] = str(aux)
    
ipc['Año'] = año
```

```python
ipc = ipc[ipc['Año'] <= 2023]
ipc = ipc[ipc['Año'] >= 2014]
ipc.reset_index(drop=True, inplace=True)
ipc
```

```python
ipc['Provincias'] = ipc['Provincias'].replace('Nacional', 'TOTAL ESPAÑA')
ipc['Total'] = ipc['Total'].apply(lambda x: re.sub(r'[^0-9]', '', x))
ipc['Total'] = ipc['Total'].astype(int)
ipc
```

```python
aux = []
for month in ipc['Periodo']:
    string = str(int(month[:2]))+'-'+month[3:]
    print (string)
    aux.append(string)
ipc['Periodo'] = aux
```

```python
names = np.unique(ipc['Provincias'])
months = ipc['Periodo'][:120]
months = months [::-1]
print (months)

provincia = []
codigo = []
ipc_dict = {x:[] for x in months}

for name in names:
    if name == 'TOTAL ESPAÑA':
        provincia.append(name)
        codigo.append('00')
    else:
        provincia.append(name[3:])
        codigo.append(name[:2])
    
    for month in months:
        value = ipc.loc[(ipc['Provincias'] == name) & (ipc['Periodo'] == month), 'Total'].iloc[0]
        while 1>0:
            if value < 90000:
                value*=10
            else:
                break
        ipc_dict[month].append(value)
```

```python
out = pd.DataFrame()
out['Codigo'] = codigo
out['Provincia'] = provincia
for month in months:
    out[month] = ipc_dict[month]
out
```

```python
out.to_csv('IPC.csv', index=False)
```

```python
pib = pd.read_csv('./Inputs/PIB_España_Trimestral.csv', header = 0, delimiter = ';')
pib = pib[(pib['Tipo de dato'] == 'Datos no ajustados de estacionalidad y calendario') & (pib['Agregados macroeconómicos: Nivel 1'] == 'Producto interior bruto a precios de mercado') & (pib['Niveles y tasas'] == 'Dato base')]
pib
```

```python
aux = []
for trim in pib['Periodo']:
    aux1 = trim[:4]
    aux2 = trim[-1]
    aux.append(aux2 + '-' + aux1)
pib['Periodo'] = aux
pib
```

```python
pib['Total'] = pib['Total'].apply(lambda x: re.sub(r'[^0-9]', '', x))
pib['Total'] = pib['Total'].astype(int)
```

```python
aux = []
for n in pib['Total']:
    if n <= 200000:
        n*=10
    aux.append(n)
pib['Total'] = aux
```

```python
outpib = {'Codigo': [], 'Provincia': []}
outpib['Codigo'].append('00')
outpib['Provincia'].append('TOTAL ESPAÑA')
trimestres = [1,1,1,2,2,2,3,3,3,4,4,4]
for year in range(2014,2024):
    for month in range (1,13):
        outpib[str(month)+'-'+str(year)] = []
        index = 4*(2023-year) + (4-trimestres[month-1])
        outpib[str(month)+'-'+str(year)].append(pib['Total'][index])
```

```python
outpib = pd.DataFrame (outpib)
outpib
```

```python
outpib.to_csv('PIB.csv', index=False)
```
