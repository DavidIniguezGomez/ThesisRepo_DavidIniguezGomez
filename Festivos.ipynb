```python
dias_semana = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
dias_mes = [31,28,31,30,31,30,31,31,30,31,30,31]
dias_bisiesto = [31,29,31,30,31,30,31,31,30,31,30,31]
meses_año = {}
for year in range (2014, 2024):
    if year == 2016 or year == 2020:
        meses_año[year] = dias_bisiesto
    else:
        meses_año[year] = dias_mes
```

```python
dia = []
mes = []
año = []
        
for year in range (2014,2024):
    for i, month in enumerate (meses_año[year]):
        for day in range (1,month+1):
            dia.append(day)
            mes.append(i+1)
            año.append(year)
#             print (day,i+1,year)

df = pd.DataFrame({'year': año,
                   'month': mes,
                   'day': dia})

fechas = pd.to_datetime(df)
```

```python
registro_dias = []
contador = 2
registro_dias.append(dias_semana[contador])
for i in range (len(dia)-1):
    contador += 1
    if contador == 7:
        contador = 0
    registro_dias.append(dias_semana[contador])

final = pd.DataFrame ({'Day': fechas, 'Day of the week': registro_dias})
    
final
```

```python
madrid_csv = pd.read_csv('./Inputs/calendario_madrid.csv',header=0,delimiter=';')
madrid_csv
# Condition to identify the row
condition = madrid_csv['Dia'] == '28/02/2016'

# Find the index of the row that satisfies the condition
index_to_insert = madrid_csv.index[condition].tolist()[0]

# New row to be added
new_row_data = {'Dia': '29/02/2016', 'Dia_semana': 'lunes', 'laborable / festivo / domingo festivo':'laborable'}

# Create a new DataFrame with the new row
madrid_csv = pd.concat([madrid_csv.iloc[:index_to_insert + 1], pd.DataFrame(new_row_data, index=[0]), madrid_csv.iloc[index_to_insert + 1:]]).reset_index(drop=True)
madrid_csv['Dia'] = pd.to_datetime(madrid_csv['Dia'], format='%d/%m/%Y')
madrid_csv=madrid_csv[(madrid_csv['Dia'].dt.year>=2014) & (madrid_csv['Dia'].dt.year<2024)].reset_index(drop=True)
madrid_csv
```

```python
autonomico = festivos[festivos['TIPO'] == 'Autonómico']
autonomico.reset_index(drop=True, inplace=True)
local = festivos[festivos['TIPO'] == 'Local']
local.reset_index(drop=True, inplace=True)
```

```python
provincias = np.unique(autonomico['PROVINCIA'])

ciudades = {x: x.upper() for x in provincias}

ciudades ['A Coruña'] = 'CORUÑA (A)'
ciudades ['Araba/Álava'] = 'VITORIA-GASTEIZ'
ciudades ['Asturias'] = 'OVIEDO'
ciudades ['Bizkaia'] = 'BILBAO'
ciudades ['Cantabria'] = 'SANTANDER'
ciudades ['Gipuzkoa'] = 'DONOSTIA-SAN SEBASTIAN'
ciudades ['Illes Balears'] = 'PALMA DE MALLORCA'
ciudades ['La Rioja'] = 'LOGROÑO'
ciudades ['Las Palmas'] = 'LAS PALMAS DE GRAN CANARIA'
ciudades ['Navarra'] = 'PAMPLONA/IRUÑA'
ciudades ['Pontevedra'] = 'VIGO'

ciudades = {value: key for key, value in ciudades.items()}

print (ciudades)
```

```python
fiestas = {x:[] for x in provincias}

# for fecha in nacional['FECHA']:
#     for name in provincias:
#         fiestas[name].append(fecha[:5])

for i in range (len(autonomico)):
    fiestas[autonomico['PROVINCIA'][i]].append(autonomico['FECHA'][i][:5])

for i in range (len(local)):
    if local['LOCALIDAD'][i] in ciudades:
        fiestas[ciudades[local['LOCALIDAD'][i]]].append(local['FECHA'][i][:5])

fiestas
```

```python
calendar = {x:[] for x in provincias}
cont = 0
for fecha in fechas:
    if madrid_csv['Tipo de Festivo'][cont] == 'Festivo nacional':
        for provincia in provincias:
            calendar[provincia].append(1)
    else:
        for provincia in provincias:
            for fiesta in fiestas[provincia]:
                flag = 0
                if int (fecha.day) == int (fiesta[:2]) and int (fecha.month) == int(fiesta[3:]):
                    flag = 1
                    break
    #                 print (provincia, fecha)
                else:
                    flag = 0
            if flag == 1:
                calendar[provincia].append(1)
            else:
                calendar[provincia].append(0)
    cont+=1
```

```python
fest = []
print("Length of final DataFrame:", len(final))  # Debug print

for i in range(len(final)):
    if final['Day of the week'][i] == 'D':
        fest.append(1)
    else:
        suma = 0
        cont = 0
        sumapob = sum(cities[str(final['Day'][i].year)][:-1])
        for name in final.columns[2:]:
            # Debug prints
#             print("Provincia:", name)
#             print("Year:", final['Day'][i].year)
#             print("Cities DataFrame:")
#             print(cities.loc[cities['Provincias'] == name])
            auxsum = final[name][i] * cities.loc[cities['Provincias'] == name][str(final['Day'][i].year)].values[0]
            suma += auxsum
            cont += 1
        fest.append(suma / sumapob)
final['sumafest'] = fest
fest
```
